## Funcation for DataFrames

```python
def two_dataframes(name_of_file='indicators_data.csv'):
    """
A Funcation takes the file name as argument and reads the file and returns two dataframes 
DataFrames: Frame1 gives the dataframe with years as a columns.Frame2 gives the dataframe with countries as a columns"""
    df = pd.read_csv(name_of_file)
    d1=pd.DataFrame(df.groupby(['Country Name', 'Series Name','Country Code','Series Code']).sum())
    d2=pd.DataFrame(d1.T)
    return d1,d2
```
